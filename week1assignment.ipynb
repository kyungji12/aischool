## 1. 언어 
 악성 댓글(악플)은 모든 커뮤니티 기반 인터넷 기업에게 고민거리다. 글로벌 사회관계망서비스(SNS) 인스타그램은 이 문제를 해결하기 위해 인공지능 기반 기술을 도입했다. 현재 한국어로 작성된 악플에도 이 기능이 적용된다.
 이 기능은 AI가 댓글이 지나치게 공격적이라고 판단하면, 경고 메세지를 표출한다. 이용자에게 '불쾌감을 줄 수 있다'는 메시지로 알려주며 이용자가 자신의 의견을 취소하고 순화할 수 있도록 기회를 준다. 

 특히 여러 연예인들의 사망을 계기로 악플로 인한 폐해가 늘어나는 가운데 포털 업체 차원의 자정 노력이 이어지고 있다. 네이버는 자체 개발한 악성 댓글 필터링 인공지능 '클린봇'을 뉴스서비스에 적용했다. 클린봇은 AI기술을 이용해 불쾌한 욕설이 포함된 댓글을 자동으로 숨겨주는 기능이다. 해당 기능을 비활성화 해도 욕설에 해당되는 단어는 여전히 가린다.


하지만 일부 문장으로도 맥락을 통해 모욕적인 뜻이 전달되어 악플을 충분히 막을 수 없다. 또한 의도적인 악플러에게는 소용이 없다. 현재 기술로 AI가 걸러내지 못하는 악플을 좀 더 정교하게 처리하는 기술이 필요하다.

 사람이 의사소통하려고 사용하는 일반적이고 효과적인 수단은 언어이고, 음성은 인간의 가장 자연스러운 의사소통 방식이다. 말의 내용과 발화자의 감정을 인식하고 의미를 이해하여, 상황에 따라 자연스러운 대화를 주고받기 위해 필요한 음성 언어 처리 기술은 인간의 자연어 발화를 컴퓨터가 자동으로 이해하고 처리하는 알고리즘을 연구하는 분야다. 단순히 음성을 인식하는 기술뿐만 아니라 인간의 섬세한 감정까지 인식하는 기술이 더욱 발전되길 바란다.

 ## 3. 이미지
브라질 상파울루 지역에 ‘프리벤트 시니어(Prevent Senior)병원’은 우리나라 인공지능 업체인 루닛(Lunit)의 ‘루닛 인사이트 CXR(폐 비정상 소견 진단 보조 인공지능 소프트웨어)’을 도입해 코로나19 의심 환자의 확진 여부를 진단하고있다. 이 소프트웨어는 흉부 엑스레이 영상을 단 몇 초 만에 분석해 폐렴 등 폐의 비정상 소견을 97~99%의 뚜어난 정확도로 검출해 주는 인공지능 알고리즘이다. 또한 의료 분야 분석 소프트웨어 업체인 레드로직스(RADLogics)는 폐의 CT영상을 관찰하고 코로나 19 확진 여부를 판단할 수 있는 인공지능 솔루션을 개발했다. 
 인공지능을 활용하면 환자를 진단하는데 시간이 얼마 안 걸리기 때문에 의사들의 격무를 줄이고 의료진 부족 사태에도 대처할 수 있을 것으로 기대된다. 

또한 AI이미지 분석 기술이 스마트팩토리에 적용되면 사람이 일일이 확인하는 것보다 오류 발생 확률을 현저히 낮추고 불량 판정률을 높일 수 있게 된다. 실제로 LG CNS의 AI이미지 분석 기술은 제조 공장에서 불량여부를 판벌하는 비전 검사 영역에 이미 적용되어있다. 

AI 이미지 인식 기술은 사람보다 더 정확해지고 있지만, AI가 그만큼 많은 양의 데이터를 학습해야한다는 문제가 있다. 사람은 몇 장의 사진만으로도 어떤 사물인지 쉽게 학습할 수 있는데, 현재의 AI는 수만 장의 사진을 통해 사진을 학습해야하기 때문이다. 사람이 정답을 알려주고 알고리즘 역시 사람이 직접 선택해야한다. 더욱이 AI가 내놓은 결과의 타당성을 최종적으로 사람이 판단해야한다. 이런 부분에서 아직은 사람이 AI의 선생님 역할을 해야하는 상황인데, AI에게 효율적으로 학습시키기 위해 나 또한 학습이 필요하다고 생각된다.

## 4. 자율주행
컴퓨팅 하드웨어 제조사인 NVIDIA는 자사의 GPU 제조 역량을 활용해 자율주행 자동차 시장에 진출하고있다. GPU는 딥러닝 기반의 인공지능 구현 시 컴퓨팅 속도 및 성능 향상을 위해 필수적으로 활용되어야 하는 핵심 부품이다. 이 부품을 기반으로 한 자율주행 기능을 소프트웨어로 구현하고, 이를 플랫폼으로 공개해 자율주행 분야의 생태계를 구축하고 있다. 

 또한 캐딜락의 슈퍼크루즈 기능은 3세대 자율주행 시스템으로 미국 내 고속도로와 일부 국도, 간선도로에서 자율주행을 할 수 있도록 지원하는 기능이다. 핸들에서 완전히 손을 뗀 채로 차량 스스로가 정밀하고 지속적인 자율주행이 가능하다.수많은 렌더링과 반복적인 측정, 시간별 도로상황까지도 모두 시스템에 포함하여 정밀 측정한 지도와 GPS, 차내 레이더와 고정밀 카메라들과 연계하여 거의 완벽한 수준의 자율주행을 선보이고있다. 
자율주행을 실현하기 위해서는 고정밀 지도 데이터가 필요한데, 이 데이터를 램프가 확실히 설치되어 있는 제한된 고속도로 내에서만 사용할 수 있기 때문에 슈퍼 크루즈 기능은 주로 고속도로에서 사용하게 될 것이다. 

우리가 흔히 알고 있는 자율 주행 기술 이외에 운전자의 의식을 점검하는 기능, 차선이탈 방지, 크루즈 주행 도중 차간 거리 간격 유지, 청각과 좌석의 진동을 이용한 입체적 경고 시스템 등 좀 더 발전한 자율 주행을 선보이며, 앞으로 ‘운전자’라는 용어 대신 ‘탑승자’라는 단어만 남게 될 것이다.
